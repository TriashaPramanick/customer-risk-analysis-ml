# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:/Users/User/OneDrive/Documents/Projects/Customer Risk Analysis\Data/Telco_churn.csv')
```

```python
df = df.drop('Churn Reason', axis=1)
```

```python
df = df.drop('CustomerID', axis=1)
```

```python
df = df.drop('Churn Label', axis=1)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
sns.countplot(x='Churn Value', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Gender', hue='Churn Value', data=df)
plt.title("Gender vs Churn")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn Value', data=df)
plt.title("Contract vs Churn")
plt.show()
```

```python
sns.boxplot(x='Churn Value', y='Monthly Charges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.boxplot(x='Churn Value', y='Tenure Months', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='Internet Service', hue='Churn Value', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Key Insights

- Customers with month-to-month contracts have higher churn
- Customers with higher monthly charges tend to churn more
- New customers (low tenure) are more likely to leave
- Certain services (like fiber internet) may have higher churn
