# Customer Risk & Engagement Analysis

## Project Objective

This project aims to analyze customer behavior and predict customer churn using machine learning techniques. The goal is to identify high-risk customers and understand the key factors influencing customer disengagement.

# What is Cutomer Churn?

Customer churn refers to the rate at which customers stop doing business with a company or cancel their subscriptions over a given period. It is a critical metric for subscription-based and recurring-revenue businesses.

# Why do businesses care about it?

1. Revenue impact: Losing customers reduces recurring revenue and increases customer acquisition costs.

2. Profitability: Retaining existing customers is significantly cheaper than acquiring new ones.

3. Brand reputation: High churn indicates poor customer satisfaction and can harm brand image.

4. Investor confidence: For recurring-revenue models, high churn signals business instability and lowers valuation.
